```python
# Environment Setup
!pip install --upgrade pip > /dev/null
!pip install requests tqdm torch torchvision numpy pandas seaborn scikit-learn > /dev/null

import os
import requests
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from tqdm.auto import tqdm
import pandas as pd
import seaborn as sns
from sklearn.feature_selection import mutual_info_classif

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", device)

################################################################################
# Step 1: Download and Prepare the dSprites Dataset (Reduced Size)
################################################################################
dsprites_url = 'https://github.com/deepmind/dsprites-dataset/blob/master/dsprites_ndarray_co1sh3sc6or40x32y32_64x64.npz?raw=true'
data_file = 'dsprites.npz'

if not os.path.exists(data_file):
    print("Downloading dSprites dataset...")
    r = requests.get(dsprites_url)
    with open(data_file, 'wb') as f:
        f.write(r.content)
    print("Download complete.")

data = np.load(data_file, allow_pickle=True, encoding='latin1')
imgs = data['imgs']
latents_values = data['latents_values']
latents_classes = data['latents_classes']

print("Full dataset loaded. Shape of imgs:", imgs.shape)

# Reduce dataset size for faster training (e.g., use first 20,000 samples)
N = 400000
imgs = imgs[:N]
latents_values = latents_values[:N]
latents_classes = latents_classes[:N]

print(f"Using subset of dataset: {N} samples")
print("Binary (0/1) images, resolution 64x64.")
print("dSprites Latent Factors: [Color=1(always), Shape(3), Scale(6), Orientation(40), posX(32), posY(32)]")
print("Focusing on factors: shape, scale, orientation, posX, posY.")

################################################################################
# Step 2: Visualize Some Sample Images
################################################################################
def show_sample_images(images, num=8):
    idxs = np.random.choice(len(images), size=num, replace=False)
    sample = images[idxs]
    grid = make_grid(torch.from_numpy(sample).unsqueeze(1).float(), nrow=4)
    plt.figure(figsize=(10,2))
    plt.imshow(grid.permute(1,2,0), cmap='gray')
    plt.title(f"Sample dSprites Images ({num} examples)")
    plt.axis('off')
    plt.show()

show_sample_images(imgs, num=8)

################################################################################
# Step 3: PyTorch Dataset and Dataloader
################################################################################
class DspritesDataset(Dataset):
    def __init__(self, images, latents_values):
        self.images = images
        self.latents = latents_values
    def __len__(self):
        return len(self.images)
    def __getitem__(self, idx):
        x = torch.from_numpy(self.images[idx]).float().unsqueeze(0)
        l = torch.from_numpy(self.latents[idx]).long()
        return x, l

dataset = DspritesDataset(imgs, latents_values)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True, num_workers=2, drop_last=True)

################################################################################
# Step 4: Define the β-VAE Model Using the Specified Architecture
################################################################################
latent_dim = 10

class BetaVAE(nn.Module):
    def __init__(self, latent_dim=10):
        super(BetaVAE, self).__init__()
        # Encoder as per the provided architecture
        # Input: 64x64x1
        # 4 x conv layers as given: (4x4 conv, stride=2)
        self.enc = nn.Sequential(
            nn.Conv2d(1, 32, 4, 2, 1), # -> 32x32
            nn.ReLU(True),
            nn.Conv2d(32, 32, 4, 2, 1), # -> 16x16
            nn.ReLU(True),
            nn.Conv2d(32, 64, 4, 2, 1), # -> 8x8
            nn.ReLU(True),
            nn.Conv2d(64, 64, 4, 2, 1), # -> 4x4
            nn.ReLU(True),
        )
        # After this, we have a 4x4x64 = 1024 features
        self.fc_mu = nn.Linear(1024, latent_dim)
        self.fc_logvar = nn.Linear(1024, latent_dim)

        # Decoder as per the provided architecture
        # Input: R^10 -> FC(256) -> FC(4x4x64)
        self.fc_dec1 = nn.Linear(latent_dim, 256)
        self.fc_dec2 = nn.Linear(256, 4*4*64)

        # Then 4 upconv layers:
        # 4x4 upconv,64 ReLU,str2 -> 8x8
        # 4x4 upconv,32 ReLU,str2 -> 16x16
        # 4x4 upconv,32 ReLU,str2 -> 32x32
        # 4x4 upconv,1 stride2 -> 64x64
        self.dec = nn.Sequential(
            nn.ConvTranspose2d(64,64,4,2,1), # -> 8x8
            nn.ReLU(True),
            nn.ConvTranspose2d(64,32,4,2,1), # ->16x16
            nn.ReLU(True),
            nn.ConvTranspose2d(32,32,4,2,1), # ->32x32
            nn.ReLU(True),
            nn.ConvTranspose2d(32,1,4,2,1),  # ->64x64
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.enc(x)
        h = h.view(h.size(0), -1) # flatten
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def decode(self, z):
        h = self.fc_dec1(z)
        h = torch.relu(h)
        h = self.fc_dec2(h)
        h = torch.relu(h)
        h = h.view(h.size(0),64,4,4)
        x_recon = self.dec(h)
        return x_recon

    def forward(self, x):
        mu, logvar = self.encode(x)
        std = torch.exp(0.5*logvar)
        eps = torch.randn_like(std)
        z = mu + eps*std
        x_recon = self.decode(z)
        return x_recon, mu, logvar

def beta_vae_loss(x, x_recon, mu, logvar, beta=1.0):
    recon_loss = nn.functional.binary_cross_entropy(x_recon, x, reduction='sum') / x.size(0)
    kl_loss = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())
    loss = recon_loss + beta*kl_loss
    return loss, recon_loss, kl_loss

################################################################################
# Step 5: Training Function
################################################################################
def train_vae(model, dataloader, epochs=10, beta=1.0, lr=1e-3):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    for ep in range(epochs):
        total_loss = 0
        total_recon = 0
        total_kl = 0
        for x, _ in dataloader:
            x = x.to(device)
            optimizer.zero_grad()
            x_recon, mu, logvar = model(x)
            loss, recon, kl = beta_vae_loss(x, x_recon, mu, logvar, beta=beta)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()*x.size(0)
            total_recon += recon.item()*x.size(0)
            total_kl += kl.item()*x.size(0)
        print(f"Epoch[{ep+1}/{epochs}] Beta={beta} | Loss: {total_loss/len(dataloader.dataset):.4f} | Recon: {total_recon/len(dataloader.dataset):.4f} | KL: {total_kl/len(dataloader.dataset):.4f}")
    return model

################################################################################
# Step 6: Train Models with Different β Values (1,5,10)
################################################################################
epochs = 5
betas = [1, 5, 10]
models = {}

for b in betas:
    print(f"\nTraining β={b} model for {epochs} epochs on {N} samples...")
    model = BetaVAE(latent_dim).to(device)
    model = train_vae(model, dataloader, epochs=epochs, beta=b)
    models[b] = model

################################################################################
# Step 7: Reconstruction Comparisons
################################################################################
def show_reconstructions(models_dict, dataloader, device):
    x, _ = next(iter(dataloader))
    x = x.to(device)
    plt.figure(figsize=(18,6))
    all_betas = sorted(models_dict.keys())
    with torch.no_grad():
        plt.subplot(2, len(all_betas)+1, 1)
        grid = make_grid(x[:8].cpu(), nrow=4)
        plt.imshow(grid.permute(1,2,0).numpy(), cmap='gray')
        plt.title("Original")
        plt.axis('off')

        for i, b in enumerate(all_betas):
            m = models_dict[b]
            m.eval()
            x_recon, _, _ = m(x[:8])
            grid = make_grid(x_recon.cpu(), nrow=4)
            plt.subplot(2, len(all_betas)+1, i+2)
            plt.imshow(grid.permute(1,2,0).numpy(), cmap='gray')
            plt.title(f"Recon β={b}")
            plt.axis('off')
    plt.suptitle("Reconstruction Comparison Across Different β Values")
    plt.show()

print("\nVisualizing Reconstructions:")
show_reconstructions(models, dataloader, device)

################################################################################
# Step 8: Latent Traversals
################################################################################
def latent_traversal(model, dataloader, device, num_dims=5, steps=5, beta_value=1):
    model.eval()
    x, _ = next(iter(dataloader))
    x = x.to(device)
    with torch.no_grad():
        mu, logvar = model.encode(x[:1])
        std = torch.exp(0.5*logvar)

        z_values = torch.linspace(-2, 2, steps).to(device)
        fig, axes = plt.subplots(num_dims, steps, figsize=(steps*2, num_dims*2))
        fig.suptitle(f"Latent Traversals (β={beta_value} model)\nVarying one latent dimension at a time", y=1.05)
        z = mu.clone()

        for dim in range(num_dims):
            for i, val in enumerate(z_values):
                z_mod = z.clone()
                z_mod[0, dim] = z[0, dim] + val*std[0, dim]
                x_rec = model.decode(z_mod)
                axes[dim, i].imshow(x_rec[0,0].cpu().numpy(), cmap='gray')
                axes[dim, i].axis('off')
            axes[dim,0].set_ylabel(f"Latent Dim {dim}", rotation=90, labelpad=10)
        plt.tight_layout()
        plt.show()

print("\nPerforming latent traversals for each β value (showing first 5 latent dims):")
for b in betas:
    latent_traversal(models[b], dataloader, device, num_dims=5, steps=7, beta_value=b)

################################################################################
# Step 9: Disentanglement Metric Evaluation (Z-diff and MIG)
################################################################################
def compute_z_diff(model, dataset, device, num_samples=2000):
    model.eval()
    idxs = np.random.choice(len(dataset), size=min(num_samples, len(dataset)), replace=False)
    X = []
    L = []
    with torch.no_grad():
        for i in idxs:
            x, l = dataset[i]
            x = x.to(device).unsqueeze(0)
            mu, logvar = model.encode(x)
            z = mu.squeeze(0).cpu().numpy()
            X.append(z)
            L.append(l.cpu().numpy())
    X = np.array(X)
    L = np.array(L)
    factors = L[:,1:]
    corrs = np.zeros((latent_dim, factors.shape[1]))
    for d in range(latent_dim):
        for f_i in range(factors.shape[1]):
            z_dim = X[:,d]
            factor = factors[:,f_i]
            z_mean = z_dim.mean()
            f_mean = factor.mean()
            num = ((z_dim - z_mean)*(factor - f_mean)).sum()
            den = np.sqrt(((z_dim - z_mean)**2).sum() * ((factor - f_mean)**2).sum())
            c = num/den if den != 0 else 0
            corrs[d,f_i] = np.abs(c)
    return corrs

def compute_mig(model, dataset, device, num_samples=10000):
    model.eval()
    idxs = np.random.choice(len(dataset), size=min(num_samples, len(dataset)), replace=False)
    X = []
    L = []
    with torch.no_grad():
        for i in idxs:
            x, l = dataset[i]
            x = x.to(device).unsqueeze(0)
            mu, logvar = model.encode(x)
            z = mu.squeeze(0).cpu().numpy()
            X.append(z)
            L.append(l.cpu().numpy())
    mus_train = np.array(X)
    ys_train = np.array(L)
    ys_train = ys_train[:,1:]

    mig_total = 0
    num_factors = ys_train.shape[1]
    for i in range(num_factors):
        factor = ys_train[:,i]
        mi = mutual_info_classif(mus_train, factor, discrete_features=False, random_state=0)
        sorted_mi = np.sort(mi)[::-1]
        if len(sorted_mi) >=2:
            mig_total += sorted_mi[0] - sorted_mi[1]
        else:
            mig_total += sorted_mi[0]
    mig_score = mig_total / num_factors
    return mig_score

def print_disentanglement_scores(model, dataset, device, name):
    corrs = compute_z_diff(model, dataset, device, num_samples=2000)
    sorted_corrs = np.sort(corrs, axis=1)
    z_diff = (sorted_corrs[:,-1] - sorted_corrs[:,-2]).mean()
    print(f"\n{name} Disentanglement Evaluation:")
    print(f"Average difference (Z-diff) between top two factor correlations: {z_diff:.4f}")

    factor_names = ["shape","scale","orientation","posX","posY"]
    df = pd.DataFrame(corrs, columns=factor_names, index=[f"Z{d}" for d in range(latent_dim)])
    print("\nCorrelation Matrix (text form):")
    print(df.to_string(float_format="%.4f"))
    print("\nCorrelation Matrix (heatmap):")
    plt.figure(figsize=(8,6))
    sns.heatmap(df, annot=True, cmap='viridis', fmt=".3f", cbar=True)
    plt.title(f"Correlation Matrix for {name}")
    plt.show()

    mig_score = compute_mig(model, dataset, device, num_samples=10000)
    print(f"{name} MIG Score: {mig_score:.4f}")
    print("Note: Higher Z-diff and MIG scores indicate better disentanglement.\n")

print("\nEvaluating disentanglement for all β values:")
for b in betas:
    print_disentanglement_scores(models[b], dataset, device, f"Beta={b}")
```
